## ***B. Create a scatterplot using 2 attributes.***

### **1. To convert categorical data to dummy variables to measure correlation.**

```python
df1 = df.copy()
df2 = pd.concat([df1, pd.get_dummies(df1['Brand'])], axis=1)
df3 = pd.concat([df2, pd.get_dummies(df2['Powerbank Capacity Size'])], axis=1)
df3.head(5)
```

### **2. To get the correlation.**

```python
df3[["Total Sold(k)", "Rock", "Aukey", "Baseus", "No Brand Stated", "Pineng", "Remax",	"TOPK",	"Xiaomi"]].corr()
```

```python
df3[["Total Sold(k)", "5000 - 10000mAh", "10001 - 20000mAh", "> 20000mAh", "No Size Stated"]].corr()
```

```python
df[["Brand", "Powerbank Capacity Size", "Total Sold(k)"]].corr()
```

From the above output, we can only view Total Sold but not Brand and Powerbank Capacity Size, because the other 2 attributes are categorical data.

### **3. To create a Scatterplot.**

```python
fig, ax = plt.subplots(figsize=(18, 10))
ax.scatter(x = df['Brand'], y = df['Total Sold(k)'])
plt.xlabel("Brand")
plt.ylabel("Total Sold(k)")

plt.show()
```
