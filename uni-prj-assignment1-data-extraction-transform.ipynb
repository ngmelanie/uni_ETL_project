# **3. Extraction of data from the links.**

To extract data from the links above, empty lists are declared first. This is so that the lists will store the scraped data. 

For this section, I have tried 2 methods to scrape the data. 

1. To scrape the product's title, I have imported BeautifulSoup to help with the scraping. BeautifulSoup will read the html and find all attributes with the id 'title'. Then, it will return the text.
2. Next, I have used the CSS Selector to get the attribute and append the value into the list. For an example, brand="Brand", the CSS Selector will go through each tag from main to div until it comes across the tag with text "Brand". Then, it will read the next div to get the next attribute value. In this case it is the 'a' tag. If nothing is found in the first row of product specification, the CSS Selector will move on to the next row. And, after going through all rows, if there is still nothing found, the system will append NaN to indicate it as a null value.

```python
# declare empty list
title_List = []

for url in links:
        driver.get(url)
        while True:
          try:
              WebDriverWait(driver, delay)
              #print ("Page is ready")
              sleep(5)
              html = driver.execute_script("return document.getElementsByTagName('html')[0].innerHTML")
              #print(html)
              soup = BeautifulSoup(html, "html.parser")


              # find the title of items
              for item_title in soup.find_all('title'):
                print(item_title.get_text())
                title_List.append(item_title.text)

              break # it will break from the loop once the specific element will be present. 
          except TimeoutException:
              print ("Loading took too much time!-Try again")      
```

# **5. Data Cleaning**

In this step of Data Cleaning, I have:-
1. Replaced certain words with another word.
2. Changed the datatype of a column.
3. Removed emojis from the texts.


In addition, the dataframe is converted to .csv file and saved for future use.

```python
# check the datatypes
df.dtypes
```

1. To replace the escape function with '>'.
2. To replace null values in the Brand column, as 'Not Stated'. This is because the product's brand is not listed in Shopee by the Seller.

```python
df['Powerbank Capacity Size'] = df['Powerbank Capacity Size'].replace("&gt; 20000mAh","> 20000mAh")
df['Brand'] = df['Brand'].replace("NaN","No Brand Stated")
df['Powerbank Capacity Size'] = df['Powerbank Capacity Size'].replace("NaN","No Size Stated")
df.head()
```

To replace the 'k' in Total Sold with thousands and then rounded off to 3 decimal places. Next, the datatype of Total Sold is changed from 'object' to 'float'.

```python
df['Total Sold(k)'] = df['Total Sold(k)'].str.extract(r'(\d+[.\d]*k*)')
df['Total Sold(k)'] = df['Total Sold(k)'].replace({'k': '*1e3'}, regex=True)
df['Total Sold(k)'] = df['Total Sold(k)'].map(pd.eval).astype(int)
df['Total Sold(k)'] = df['Total Sold(k)'].div(1e3).round(3)
df.head()
```

```python
df.dtypes
```

To remove emojis from the Product Title.

```python
def deEmojify(inputString):
    return inputString.encode('ascii', 'ignore').decode('ascii')

row = df['Product Title']
for n in row:
  df["Product Title"] = df["Product Title"].apply(lambda s: deEmojify(s))
  df[['Product Title','Product Title']].iloc[12]

df.head()
```

# **6. Data Binning**

In this step of Data Binning, I have:-
1. Created a column, Product Sales.
2. Divided the total sold values into discrete intervals.
3. Create a label for each intervals, which are Lowest Selling, Medium Selling, and Highest Selling.

The column 'Product Sales' indicates the level of popularity in sales of a product. Lowest Selling, meaning out of all the extracted data, it has the lowest total of products sold. And in contrast, Highest Selling, will indicate that a product has sold the most, out of all extracted products.

```python
# give grade/level to specific range of number
df1 = df.copy()
bins = [0, 38, 76, 114]
labels =["Lowest Selling", "Medium Selling", "Highest Selling"]
#cut() Method: Bin Values into Discrete Intervals
df1['Product Sales'] = pd.cut(df1['Total Sold(k)'], bins, labels=labels)
df1.head()
```
